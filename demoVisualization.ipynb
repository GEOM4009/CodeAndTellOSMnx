```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Define the area of interest
place_name = "Ciutat Vella, Barcelona"

# Retrieve the boundary geometry of the specified place and plot it
area = ox.geocode_to_gdf(place_name)
area.plot()
```

```python
# Specify tags to filter the desired features (in this case, buildings)
tags = {"building": True}

# Retrieve building geometries within the defined area using the specified tags
buildings = ox.features_from_place(place_name, tags)

# Create a figure and axis for the plot
f, ax = plt.subplots(1, 1, figsize=(7, 11))

# Plot the building geometries in a specific color
buildings.plot(ax=ax, color='tomato')

# Remove axis labels for a cleaner visualization
plt.axis('off')

# Set the background color of the plot to black
plt.gcf().set_facecolor('black')

# Adjust the layout to prevent clipping of titles
plt.subplots_adjust(top=0.82)

# Add a main title to the plot with customized font settings
plt.suptitle("Ciutat Vella", fontsize=15, fontweight='bold', color='tomato', y=0.8)

# Add a subtitle to the plot with smaller font size and different positioning
plt.title("Barcelona", fontsize=9, fontweight='light', color='tomato', y=-0.1)
```
